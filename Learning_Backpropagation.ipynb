```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])
```

```python
def initialise_parametres(layer_dimensions):
  parametres = {}
  L = len(layer_dimensions)
  for l in range(1 , L):
      parametres['W' + str(l)] = np.ones((layer_dimensions[l-1] , layer_dimensions[l]))*0.1
      parametres['b' + str(l)] = np.ones((layer_dimensions[l] , 1))

  return parametres
```

```python
initialise_parametres([2 , 2 , 1])
```

```python
def linear_forward(A_prev , wl , bl):
  z = np.dot(wl . T , A_prev) + bl
  return z
```

```python
def linear_propagation(X , parametres) :
  A = X ;
  L = len(parametres)//2
  for l in range(1 , L+1) :
    A_prev = A
    wl = parametres['W' + str(l)]
    bl = parametres['b' + str(l)]

    A = linear_forward(A_prev , wl , bl)
  return A , A_prev
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]
```

```python
parametres = initialise_parametres([2 ,2 , 1])
y_hat , A1 = linear_propagation(X , parametres)
```

```python
def update_parametres(parametres, y, y_hat, A1, X):
    parametres['W2'][0][0] = parametres['W2'][0][0] + (0.001 * 2 * (y - y_hat) * A1[0][0])
    parametres['W2'][1][0] = parametres['W2'][1][0] + (0.001 * 2 * (y - y_hat) * A1[1][0])
    parametres['b2'][0][0] = parametres['b2'][0][0] + (0.001 * 2 * (y - y_hat))

    parametres['W1'][0][0] = parametres['W1'][0][0] + (0.001 * 2 * (y - y_hat) * parametres['W2'][0][0] * X[0][0])
    parametres['W1'][0][1] = parametres['W1'][0][1] + (0.001 * 2 * (y - y_hat) * parametres['W2'][0][0] * X[1][0])
    parametres['b1'][0][0] = parametres['b1'][0][0] + (0.001 * 2 * (y - y_hat) * parametres['W2'][0][0])

    parametres['W1'][1][0] = parametres['W1'][1][0] + (0.001 * 2 * (y - y_hat) * parametres['W2'][1][0] * X[0][0])
    parametres['W1'][1][1] = parametres['W1'][1][1] + (0.001 * 2 * (y - y_hat) * parametres['W2'][1][0] * X[1][0])
    parametres['b1'][1][0] = parametres['b1'][1][0] + (0.001 * 2 * (y - y_hat) * parametres['W2'][1][0])
```

```python
parametres = initialise_parametres([2 ,2 , 1])
y_hat , A1 = linear_propagation(X , parametres)
y_hat = y_hat[0][0]
update_parametres(parametres , y , y_hat , A1 , X)
```
