```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
def Perceptron(X , y):
  X = np.insert(X,0,1,axis = 1)
  weights = np.ones(X.shape[1])
  lr = 0.1
  for i in range(1000):
    j = np.random.randint(0,100)
    y_pred = step(np.dot(X[j],weights))
    weights = weights + lr*(y[j] - y_pred)*X[j]

  return weights[0] , weights[1 : ]
```

```python
def step(x):
  return 1 if x > 0 else 0
```

```python
intercept, coef = Perceptron(X , y)
print(intercept)
print(coef)
```

```python
m = -(coef[0] / coef[1])
c = -(intercept / coef[1])
```

```python
x_input = np.linspace(-3 , 3 , 100)
y_input = (m*x_input) + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
def predict(student , intercept , coef):
  y = np.dot(student , coef) + intercept
  prediction = step(y)
  if(prediction == 0) :
    print("The Student is not Placed")
  else :
    print("The Student is Placed")
```

```python
CGPA = 0.19924745
IQ = -0.11697552
student = [CGPA , IQ]
predict(student , intercept , coef)
```
